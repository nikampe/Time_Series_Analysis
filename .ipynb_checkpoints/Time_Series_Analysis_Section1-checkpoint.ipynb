## Example: White Noise

```python
from statsmodels.graphics.tsaplots import plot_acf 
from statsmodels.tsa.stattools import acf
```

```python
np.random.seed(453)
n = 800
mu = 0 
sigma = 1
wt = np.random.normal(mu, sigma, size = n)
plt.figure(figsize=(16,6))
plt.plot(wt,linestyle=':', marker='o', color='k')
plt.xlabel("t")
plt.ylabel("$w_t$")
plt.title("White Noise")
plt.grid(False)
```

```python
fig = plot_acf(wt,lags = 40)
plt.xlabel("lag")
plt.ylabel("ACF")
plt.title("ACF of Wite Noise, n = " + str(n))
```

```python
wt_acf = acf(wt)
for h in range(0,10):
    print("gamma("+str(h)+")="+str(wt_acf[h]))
```

## Example: White Noise with Drift

```python
np.random.seed(453)
n = 800
mu = 0 
sigma = 1
wt = np.random.normal(mu, sigma, size = n)
xt = [0.01*t + wt[t] for t in range(0,n)]

plt.figure(figsize=(16,6))
plt.plot(xt,linestyle=':', marker='o', color='k')
plt.xlabel("t")
plt.ylabel("$x_t$")
plt.title("White Noise with Drift")
plt.grid(False)
```

```python
fig = plot_acf(xt,lags = 40)
plt.xlabel("lag")
plt.ylabel("ACF")
plt.title("ACF of Wite Noise with Drift, n = " + str(n))
```

## Other Examples

## Ex. 1

```python
n = 1100
xt = range(1,n+1)
plt.figure(figsize=(16,6))
plt.plot(xt,linestyle=':', marker='o', color='k')
plt.xlabel("t")
plt.ylabel("$x_t$")
plt.title("$x_t=t$")
fig = plot_acf(xt,lags = 40,alpha=0.01)
plt.title("ACF of $x_t=t$, $n =$ " + str(n))
```

## Ex. 2

```python
n = 800
wt = np.random.normal(loc=0, scale=1, size=n)
xt = wt[1:n] + 0.9*wt[0:(n-1)]
```

```python
fig = plot_acf(xt,lags = 40,alpha=0.01)
plt.title("ACF of $x_t=w_t + 0.9 w_{t-1}$")
```
