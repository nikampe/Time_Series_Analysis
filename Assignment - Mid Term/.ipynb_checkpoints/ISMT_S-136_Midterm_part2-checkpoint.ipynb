More information about the dataset can be found at http://www.sidc.be/silso/datafiles. We will build a time series model for the number of sunspots using 1700-2008 measurements and evaluate the model performance using the last 10 years of observations, 2009-2018. 

(a) Load the time series data from the ‘sunspots_yearly.csv’ file and save data as a time series object. Call it $x_t$. Check how many observations the dataset has and verify that there are no missing values. Next, plot the time series. Make sure that the x-axis’ tick values are years (i.e. not just indexes 1,2,..). Add the title "Sq. root of yearly number of sunspots from 1700 to 2018," x-label "Year," and y-label "$y_t = \sqrt{x_t}$." Observe that $Var[y_t]$ is nearly constant, so we will want to build a model for $y_t$, not $x_t$.   
(c) Split the data set into two parts: training and testing using 1700-2008 and 2009-2018 observations, respectively. You should obtain the testing time series of length 10. At this point, use only the training data set for your analysis and fitting the model - use the testing time series to only evaluate the model performance.   
(d) Display the ACF and PACF functions of training $y_t$ (i.e using 1700-2008 for computing ACF and PACF) for lags 0 through 60. Let's use ARMA($p$,$q$) model. Decide what $p$ and $q$ you want to use. Explain.

SOLUTION:

```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import statsmodels.tsa.api as smt

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf 
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.tsa.arima.model import ARIMA

from sklearn.metrics import mean_squared_error

import warnings
warnings.filterwarnings('ignore')
```

(a)

```python
df = pd.read_csv('sunspots_yearly.csv', parse_dates=['year'],index_col='year')
df.rename(columns={"number": "x_t"}, inplace = True)
print('Number of observations:', int(df.count()))
print('Data set included NAs:', df.isnull().values.any())
```

```python
plt.plot(df.index, df['x_t'])
plt.title('Yearly number of sunspots from 1700 to 2018', size = 14)
plt.xlabel('Year')
plt.ylabel('$x_t$')
plt.show()
```

(b)

```python
df['y_t'] = np.sqrt(df['x_t'])
plt.plot(df.index, df['y_t'])
plt.title('Sq. root of yearly number of sunspots from 1700 to 2018', size = 14)
plt.xlabel('Year')
plt.ylabel('$y_t = \sqrt{x_t}$')
plt.show()
```

The variance of $y_t$ over time is nearly constant.

(c)

```python
year = 2008
df_train = df[df.index.year <= year]['y_t']
df_test = df[df.index.year > year]['y_t']
print("Length of training data:", len(df_train))
print("Length of test data:", len(df_test))
```

(d)

```python
h_max = 60

# ACF
plot_acf(df_train, lags = h_max)
plt.ylim(-1.1, 1.1)
plt.xlabel("Lags")
plt.ylabel("ACF")
plt.title('ACF of $y_t = \sqrt{x_t}$',size = 14)
plt.grid(True)
plt.show()

# PACF
plot_pacf(df_train, lags = h_max)
plt.ylim(-1.1, 1.1)
plt.xlabel("Lags")
plt.ylabel("PACF")
plt.title('PACF of $y_t = \sqrt{x_t}$',size = 14)
plt.grid(True)
plt.show()
```

Based on the ACF and PACF plots, the time series $y_t = \sqrt{x_t}$ could be fit by an ARMA(p,q) model with the orders $p = 9$ of the autoregressive part of the ARMA(p,q) model (based on the out-of-bounds PACFs with increasing lags; in-bound PACFs or PACFs close to the negative/positive bound are assumed to be insignificant) and $q = 28$ for the moving average part of the ARMA(p,q) model (based on the out-of-bounds ACFs with increasing lags; in-bound ACFs or ACFs close to the negative/positive bound are assumed to be insignificant).

(e)

```python
# Model fit
p = 1 #p validated in part (h)
q = 31 #q validated in part (h)
model = ARIMA(df_train, order = (p,0,q), enforce_stationarity = False, enforce_invertibility = False)
model_fit = model.fit()
print(model_fit.summary())

# Causality and invertibility check
phi = model_fit.params[1]
theta = model_fit.params[2:len(model_fit.params)-1]
ar = np.r_[1, -phi]
ma = np.r_[1, theta]
sigma = 1    
arma_process_AR = smt.ArmaProcess(ar, ma)
print('\nProcess is causal:', arma_process_AR.isstationary)
print('Process is invertible:', arma_process_AR.isinvertible)
```

From the results above, based on the evaluated MSE-minimizing AR- and MA-orders, the process is causal but not invertible. This is mainly due to the low order of the AR-part and the high order of the MA-part, which leads to roots inside the unit circle for the AR-part, but roots outside the unit circle for the MA-part.

(f)

```python
year = 2008
m = 10
n = len(df_train)

# Forecast Plot
yt_forecasted = model_fit.predict(start = n, end = n+m-1)
xt_forecasted = np.square(yt_forecasted)
xt_orginal = df[df.index.year <= year]['x_t']
print('10 year forecasted values:\n', xt_forecasted.to_numpy())
plt.figure(figsize=(16,6))
plt.plot(range(1,n+1), xt_orginal, linestyle=':', marker='o', color='blue', label = "$y_t$")
plt.axvline(x = n, linestyle=':', color='k')
plt.plot(range(n+1,n+m+1), xt_forecasted, linestyle=':', marker='o',color='red',label = "forecasted $y_t$")
plt.title('$x_t$ and forecasted $x_t$ for ARMA(1,31) model', size = 14) #q and p validated in part (h)
plt.xlabel('t')
plt.ylabel('$x_t$')
plt.legend()
plt.show()
```

(g)

```python
mse = mean_squared_error(df_test, yt_forecasted, squared = True)
print('Mean Squared Error:', mse)
```

(h)

```python
# MSE validation across all out-of-bounds AR- and MA-orders

m = 10
n = len(df_train)

p_min = 0
p_max = 18 #highest out-of-bounds/significant partial autocorrelation in PACF
q_min = 0
q_max = 38 #highest out-of-bounds/significant autocorrelation value in ACF

mse_df = pd.DataFrame({'p': [], 'q': [], 'MSE': []})
columns = list(mse_df)
data = []

for p in range(p_min, p_max):
    for q in range(q_min, q_max):
        model = ARIMA(df_train, order = (p,0,q), enforce_stationarity = False, enforce_invertibility = False)
        model_fit = model.fit()
        yt_forecasted = model_fit.predict(start = n, end = n+m-1)
        mse = mean_squared_error(df_test, yt_forecasted, squared = True)
        values = [int(p), int(q), mse]
        zipped = zip(columns, values)
        dictionary = dict(zipped)
        mse_df.append([p, q, mse])
        data.append(dictionary)
mse_df = mse_df.append(data, True)
pd.options.display.float_format = '{:.2f}'.format
print(mse_df)
print('\nSmallest MSE:')
print(mse_df[mse_df['MSE'] == mse_df['MSE'].min()])
```

## Problem 2 (10 points)
In this problem we will consider a time series $x_t$ with a trend and use differencing techniques to transfer $x_t$ to a stationary time series. File 'xt_midterm.txt' contains observations that were generated by ARMA process with a trend added to it.

(a) Load 'xt_midterm.txt' – lets call it $x_t$. Plot the time series, and observe that the time series is not stationary because of the trend. Does the trend appear to be linear, quadratic, etc.?

(c) Reserve the last 10 data points for comparison - do not use them for the model analysis and model fit. Plot ACF and PACF of $\nabla x_t$. Please use method='ywm' in plot_pacf(), which corresponds to "yule walker without bias correction." Identify $p$ and $q$.

(e) Forecast the last 10 observations of $\nabla x_t$ and then compute the corresponding forecasts of $x_t$. Please be careful here with indexes: $\nabla x_t$ has one data point less than $x_t$ because of the differencing.  
(f) Plot the original time series $x_t$ and the forecasts of the last 10 observations in one graph. Make sure to add a legend in order to differentiate between observed $x_t$ and the forecast of 10 last data points.   
(g) Compute the mean squared error (MSE) for the last 10 predictions of $x_t$.   
(h) Finally, revisit your model and make sure your selection indeed corresponds to minimal MSE. SOLUTION:

(a)

```python
xt = np.loadtxt("xt_midterm.txt")
plt.plot(xt)
plt.title('Non-stationary time series $x_t$ with trend', size = 14)
plt.xlabel('Time')
plt.ylabel('$y_t$')
plt.grid(True)
plt.show()
```

The time series $x_t$ seems to include a positive linear trend wihich is observable by linearly increasing values of $x_t$ in increasing time $t$.

(b)

```python
df = pd.DataFrame(xt, columns = {'x(t)'})
for i in range(len(df)):
    if i == 0:
        df.loc[i, 'B_x(t)'] = np.nan
    elif i == 1:
        df.loc[i, 'BB_x(t)'] = np.nan
        df.loc[i, 'B_x(t)'] = df.loc[i, 'x(t)'] - df.loc[i-1, 'x(t)']
    elif i == 2:
        df.loc[i, 'BBB_x(t)'] = np.nan
        df.loc[i, 'BB_x(t)'] = df.loc[i, 'x(t)'] - df.loc[i-2, 'x(t)']
        df.loc[i, 'B_x(t)'] = df.loc[i, 'x(t)'] - df.loc[i-1, 'x(t)']
    else:
        df.loc[i, 'BBB_x(t)'] = df.loc[i, 'x(t)'] - df.loc[i-3, 'x(t)']
        df.loc[i, 'BB_x(t)'] = df.loc[i, 'x(t)'] - df.loc[i-2, 'x(t)']
        df.loc[i, 'B_x(t)'] = df.loc[i, 'x(t)'] - df.loc[i-1, 'x(t)']
plt.figure(figsize=(16,6))
plt.plot(df['BBB_x(t)'], label = '$\\nabla x_t$')
plt.plot(df['BB_x(t)'], label = '$\\nabla^2 x_t$')
plt.plot(df['B_x(t)'], label = '$\\nabla^3 x_t$')
plt.title('$\\nabla x_t$, $\\nabla^2 x_t$ and $\\nabla^3 x_t$ over time', size = 14)
plt.xlabel('$t$')
plt.ylabel('$x_t$')
plt.legend()
plt.show()
```

Based on the plot of the first-, second- and third-order difference of $x_t$, it can be concluded that the smallest difference order leading to weak stationarity is $d = 1$. This can be observed in the plot in which all differences show a variance which is close to constant over time and an expected value which is close to 0. This conclusion is further supported by the assumption that positive linear trends can be eliminated by the first order difference in order to convert the time series into stationarity.

(c)

```python
n = 10
h_max = 60

df_train = df.drop(df.tail(n).index)
df_train = df_train['B_x(t)']
df_train = df_train.drop([0], axis = 0)

# ACF
plot_acf(df_train, missing = 'drop', lags = h_max)
plt.ylim(-1.1, 1.1)
plt.xlabel("Lags")
plt.ylabel("ACF")
plt.title('ACF of $\\nabla x_t$',size = 14)
plt.grid(True)
plt.show()

# PACF
plot_pacf(df_train, method = 'ywm', lags = h_max)
plt.ylim(-1.1, 1.1)
plt.xlabel("Lags")
plt.ylabel("PACF")
plt.title('PACF of $\\nabla x_t$', size = 14)
plt.grid(True)
plt.show()
```

Based on the ACF and PACF plots, the first-order differenced time series $\\nabla x_t$ could be fit by an ARMA(p,q) model with the orders $p = 17$ of the autoregressive part of the ARMA(p,q) model (based on the out-of-bounds PACFs with increasing lags; in-bound PACFs or PACFs close to the negative/positive bound are assumed to be insignificant) and $q = 12$ for the moving average part of the ARMA(p,q) model (based on the out-of-bounds ACFs with increasing lags; in-bound ACFs or ACFs close to the negative/positive bound are assumed to be insignificant).

(d)

```python
# Model fit
p = 3 # p validated in part (h)
q = 0 # q validated in part (h)
model = ARIMA(df_train, order = (p,0,q), enforce_stationarity = False, enforce_invertibility = False)
model_fit = model.fit()
print(model_fit.summary())
```

(e)

```python
m = 10
n = len(df_train)

xt_forecasted = model_fit.predict(start = n, end = n+m-1)
print('10 period forecasted values:\n', xt_forecasted.to_numpy())
```

(f)

```python
m = 10
n = len(df_train)

# Forecasted plots
xt_forecasted = model_fit.predict(start = n, end = n+m-1)
plt.figure(figsize=(16,6))
plt.plot(range(1,n+1), df_train, linestyle=':', marker='o',color='blue',label = "$\\nabla x_t$")
plt.axvline(x = n,linestyle=':', color='k')
plt.plot(range(n+1,n+m+1), xt_forecasted, linestyle=':', marker='o',color='red',label = "forecasted $\\nabla x_t$")
plt.title('$\\nabla x_t$, fitted $\\nabla x_t$ and forecasted $\\nabla x_t$ for ARMA(3,0) model', size = 14) #p and q validated in (h)
plt.xlabel('t')
plt.ylabel('$\\nabla x_t$')
plt.legend()
plt.show()
```

(g)

```python
n = 10

df_test = df.tail(n)
df_test = df_test['B_x(t)']

mse = mean_squared_error(df_test, xt_forecasted, squared = True)
print('Mean Squared Error:', mse)
```

(h)

```python
# MSE validation across all out-of-bounds AR- and MA-orders

m = 10
n = len(df_train)

p_min = 0
p_max = 17 #highest out-of-bounds/significant partial autocorrelation in PACF
q_min = 0
q_max = 12 #highest out-of-bounds/significant autocorrelation in ACF

mse_df = pd.DataFrame({'p': [], 'q': [], 'MSE': []})
columns = list(mse_df)
data = []

for p in range(p_min, p_max):
    for q in range(q_min, q_max):
        model = ARIMA(df_train, order = (p,0,q), enforce_stationarity = False, enforce_invertibility = False)
        model_fit = model.fit()
        xt_forecasted = model_fit.predict(start = n, end = n+m-1)
        mse = mean_squared_error(df_test, xt_forecasted, squared = True)
        values = [int(p), int(q), mse]
        zipped = zip(columns, values)
        dictionary = dict(zipped)
        mse_df.append([p, q, mse])
        data.append(dictionary)
mse_df = mse_df.append(data, True)
pd.options.display.float_format = '{:.2f}'.format
print('\nSmallest MSE:')
print(mse_df[mse_df['MSE'] == mse_df['MSE'].min()])
```

The ARMA(3,0) autoregressive and moving average orders which minimize the mean squared error (MSE) are $p = 3$ and $q = 0$.
