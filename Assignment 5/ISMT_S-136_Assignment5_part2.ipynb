## Problem 1 (10 points)
In this problem, we study Partial ACF (PACF) of AR($p$) and MA($q$) processes. In each case below please plot the theoretical PACF (using an object created with statsmodels.tsa and .pacf() method) for lags $h=0,1,3,\ldots,30$ and also sample PACF based on 300 runs for the same range of lags. Make sure to indicate what case is being considered in titles. Whenever a time series is generated, please set the following seed: np.random.seed(345).

(a) AR(1) with $\phi_1 = 0.85$.   
(b) AR(2) with $\phi_1 = 1.2$ and $\phi_2 = -0.4$.   
(c) AR(3) with $\phi_1 = 1.1$, $\phi_2 = -0.6$, and $\phi_3=0.3$.   
(d) MA(3) with $\theta_1=1.4$, $\theta_2 = 0.5$, and $\theta_3=0.1$.    

SOLUTION:

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.tsa.api as smt

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf 
from statsmodels.tsa.stattools import acf, pacf
```

(a)

```python
# AR(1) with phi(1) = 0.85

## Theoretical PACF
h_max = 30
phi = np.array([0.85])
theta = np.array([0.])
ar = np.r_[1, -phi]
ma = np.r_[1, theta]
sigma = 1
ar_1 = smt.ArmaProcess(ar, ma)
theoretical_pacf = ar_1.pacf(h_max)

## Sample PACF
np.random.seed(345)
n = 300
xt = smt.arma_generate_sample(ar=ar, ma=ma, scale=sigma, nsample=n)

## Theoretical PACF Plot
plt.plot(theoretical_pacf, linestyle='None', marker='o')
plt.ylim(-0.1, 1.1)
plt.xlabel("Lags")
plt.ylabel("Theoretical PACF")
plt.title('Theoretical PACF of AR(1) with $\phi_1=$'+str(phi[0]),size=16)
plt.grid(True)
plt.show()

## Sample PACF Plot
plot_pacf(xt, lags=h_max)
plt.ylim(-0.1, 1.1)
plt.xlabel("Lags")
plt.ylabel("Sample PACF")
plt.title('Sample PACF of AR(1) with $\phi_1=$'+str(phi[0]),size=16)
plt.grid(True)
plt.show()
```

## Problem 2 (10 points)
In this problem we will identify the order $p$ of AR($p$) based on a sample run. File 'xt.txt' contains observations that were generated by AR process. The data, however, contain missing values. Your task is to fill the missing values first using three different approaches (a),(b), and (c) - please see below - then plot the time series in each case. Finally, we need to estimate the PACF, plot it, and identify the order $p$ - please repeat this for each cases:

(a) forward fill: fill in the missing values with the preceding observation, that is, if $x_t$ is missing, replace it with $x_{t-1}$     
(b) backward fill: fill in the missing values with the following observation, that is, if $x_t$ is missing, replace it with $x_{t+1}$     
(a) fill NaNs with the mean: fill in the missing values with the mean of observed values

SOLUTION:

```python
# Case 1
xt_1 = np.loadtxt("xt.txt")
for i in range(0,len(xt_1)):
    if np.isnan(xt_1[i]) == False:
        None
    else:
        xt_1[i] = xt_1[i-1]
# print("\n Modified time series (Case 1): \n", xt_1)

## Time Series Plot
plt.figure(figsize=(16,4))
plt.plot(xt_1)
plt.ylabel("$x_t$",fontsize=14)
plt.title("Case 1: Forward Fill",fontsize=16)
plt.grid(True)

## PACF Plot
h_max = 30
plot_pacf(xt_1, lags=h_max)
plt.ylim(-1.1, 1.1)
plt.xlabel("Lags")
plt.ylabel("Sample PACF")
plt.title('Sample PACF', size=16)
plt.grid(True)
plt.show()
```

Order p of the associated AR(p) process: 3 (all further partial autocorrelations are inside the normal bounds and therefore are not significant)

```python
# Case 2
xt_2 = np.loadtxt("xt.txt")
for i in range(0,len(xt_2)):
    if np.isnan(xt_2[i]) == False:
        None
    else:
        xt_2[i] = xt_2[i+1]
# print("\n Modified time series (Case 2): \n", xt_2)

## Time Series Plot
plt.figure(figsize=(16,4))
plt.plot(xt_2)
plt.ylabel("$x_t$",fontsize=14)
plt.title("Case 2: Backward Fill",fontsize=16)
plt.grid(True)

## PACF Plot
h_max = 30
plot_pacf(xt_2, lags=h_max)
plt.ylim(-1.1, 1.1)
plt.xlabel("Lags")
plt.ylabel("Sample PACF")
plt.title('Sample PACF', size=16)
plt.grid(True)
plt.show()
```

Order p of the associated AR(p) process: 3 (all further partial autocorrelations are inside the normal bounds and therefore are not significant)

```python
# Case 3
xt_3 = np.loadtxt("xt.txt")
for i in range(0,len(xt_3)):
    if np.isnan(xt_3[i]) == False:
        None
    else:
        xt_3[i] = np.nanmean(xt_3)
# print("\n Modified time series (Case 3): \n", xt_3)

## Time Series Plot
plt.figure(figsize=(16,4))
plt.plot(xt_3)
plt.ylabel("$x_t$",fontsize=14)
plt.title("Case 3: Mean Fill",fontsize=16)
plt.grid(True)

## PACF Plot
h_max = 30
plot_pacf(xt_3, lags=h_max)
plt.ylim(-1.1, 1.1)
plt.xlabel("Lags")
plt.ylabel("Sample PACF")
plt.title('Sample PACF', size=16)
plt.grid(True)
plt.show()
```

Order p of the associated AR(p) process: 6 (all further partial autocorrelations are inside the normal bounds and therefore are not significant)
