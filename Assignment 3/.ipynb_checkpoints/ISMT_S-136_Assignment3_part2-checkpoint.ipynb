## Problem 1 (10 points)
In this problem, we will numerically investigate stationarity of AR(1) process. Please consider the following AR(1) time series   

$$(1-\phi B)x_t=w_t,$$   
where  $w_t$ is Gaussian White Noise with $Var(w_t)=1$. Here, $B$ denotes the backshift operator.

(a) For $\phi=0.5$, generate one realization $x_1,x_2,\dots,x_{400}$ of the AR(1) process. Plot observations versus time $t=1,2,\ldots,400$, indicate the value of $\phi$ in the title. Plot the estimate of the ACF. Comment on what you observe and specify whether (how) the sample ACF differs form the ACF of AR(1) with $\phi=0.5$.

(b) Repeat (a) for $\phi=-0.5$. Comment.

(c) Generate $M=500$ realizations of AR(1) with $\phi=0.5$ for $t=1,2,\ldots,40$. Plot all $M$ realizations in one graph, indicate the value of $\phi$ in the title, and make sure to connect observations that belong to the same realization. Next, plot the three histograms of observed $x_{20}$, $x_{30}$, and $x_{40}$, please make sure to properly label each histogram, for example, by adding titles $x_{20}$ etc. Comment on whether you think the histograms represent same distribution. Also, please answer whether the distributions of $x_{20}$, $x_{30}$, and $x_{40}$ are expected to be same in case of AR(1) with $\phi=0.5$.  

(d) Repeat (c) for $\phi=1.2$. Answer the questions.

SOLUTION:

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.tsa.api as smt

from statsmodels.graphics.tsaplots import plot_acf 
from statsmodels.tsa.stattools import acf
```

(a)

```python
# AR(1) with phi = 0.5
np.random.seed(12345678)

n = int(400)
phi = np.array([0.5])
theta = np.array([0])
AR = np.r_[1, -phi]
MA = np.r_[1, theta]
sigma = 1
xt = smt.arma_generate_sample(ar = AR, ma = MA, scale = sigma, nsample = n)

fig = plot_acf(xt,lags = 20)
plt.xlabel("Lags")
plt.ylabel("Sample ACF")
plt.title("Sample ACF of AR(1) with $\phi=0.5$")
```

An AR(1) process with a parameter of phi = 0.5 has a monotonically decreasing ACF with autocorrelations greater than zero, while the Sample ACF of an AR(1) process with the same parameter phi = 0.5 is not strictly decreasing over the lags and also has autocorellations smaller than zero. Furthermore, the number of values inside the bounds of the Sample ACF is higher over the same lags than for the ACF.

(b)

```python
# AR(1) with phi = -0.5
np.random.seed(12345678)

n = int(400)
phi = np.array([-0.5])
theta = np.array([0])
AR = np.r_[1, -phi]
MA = np.r_[1, theta]
sigma = 1
xt = smt.arma_generate_sample(ar = AR, ma = MA, scale = sigma, nsample = n)

fig = plot_acf(xt,lags = 20)
plt.xlabel("lag")
plt.ylabel("Sample ACF")
plt.title("Sample ACF of AR(1) with $\phi=0.5$")
```

An AR(1) process with a parameter of phi = -0.5 has an ACF with autocorrelations that are monotonically decreasing in absolute values over lags and values jumping from negative to positive, while the Sample ACF of an AR(1) process with the same parameter phi = -0.5 is not strictly decreasing in absolute values over the lags and also has autocorellations of the same sign on consecutive lags. Furthermore, the number of values inside the bounds of the Sample ACF is higher over the same lags than for the ACF.

(c)

```python
import pandas as pd

np.random.seed(12345678)

df_AR = pd.DataFrame(index = range(1,501), columns = range(1,41))

M = 500
n = 40
phi = np.array([0.5])
theta = np.array([0])
sigma = 1
AR = np.r_[1, -phi]
MA = np.r_[1, theta]

for i in range(M):                  
    xt = smt.arma_generate_sample(ar = AR, ma = MA, scale = sigma, nsample = n)
    df_AR.loc[i+1] = xt

plt.plot(df_AR)
plt.ylabel('$x_t$')
plt.xlabel('Realizations')
plt.title('500 realizations of AR(1) with $\phi = 0.5$')
plt.show
```

```python
fig, (h1, h2, h3) = plt.subplots(nrows = 1, ncols = 3, sharex = False, figsize = (15,3))
fig.suptitle('AR(1) with $\phi = 0.5$',y = 1.08, size = 16)
plt.subplots_adjust(wspace = 0.5)

h1.hist(df_AR.loc[:,20], linestyle="None", color='k')
h1.set_title("x_20")
h1.set_xlabel("$x_t$")
h1.set_ylabel("Count")

h2.hist(df_AR.loc[:,30], linestyle="None", color='k')
h2.set_title("x_30")
h2.set_xlabel("$x_t$")
h2.set_ylabel("Count")

h3.hist(df_AR.loc[:,40], linestyle="None", color='k')
h3.set_title("x_40")
h3.set_xlabel("$x_t$")
h3.set_ylabel("Count")

plt.show
```

The distributions of the realizations of x_20, x_30 and x_40 with phi = 0.5 do not seem to come from the same distribution because the histograms seem to have different values for the mean, variance, kurtosis and skewness. Therefore it can be assumened from the samples that they do not result from the same underlying distribution.
