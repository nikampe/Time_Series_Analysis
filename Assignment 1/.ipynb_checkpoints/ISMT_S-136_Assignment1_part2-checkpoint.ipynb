## Problem 1 (10 points)
Load the Dow Jones Industrial Average (DJIA) from djia.csv to a data frame. Display the first 10 rows using .head(), check how many observations the dataset has, and verify that there are no missing values. Let's now focus on the Close value of the index ('Close' column):

(a) Use matplotlib.pyplot to plot the Close index versus time. Add title "The Dow Jones Industrial Average (DJIA) from April 20, 2006 to April 20, 2016," x-label "Time," and y-label "DJIA."   
Does the time series appear to be stationary? Please justify.

(b) Compute daily returns that are based on the close DJIA. Use .diff(). Please note that the return, namely $r_t$, for a time series $x_t$ is defined as follows: $r_t\doteq \frac{x_t-x_{t-1}}{x_{t-1}}$. It is, however, a normal practice to approximate it with $\ln(1+r_t)=\ln(\frac{x_t}{x_{t-1}})=\ln(x_t)-\ln(x_{t-1})$. So, let's use this approximation.   
Plot the return similarly to (a) changing the title to "The daily returns of the Dow Jones Industrial Average (DJIA) from April 20, 2006 to April 20, 2016" and y-label to "DJIA Returns."     
Does the time series of the daily returns appear to be stationary? Please justify.

SOLUTION:

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('djia.csv', parse_dates=['date'],index_col='date')
df.head(n=10)
```

```python
df.count()
```

```python
df.isnull().values.any()
```

(a)

```python
plt.plot(df.index, df['Close'])
plt.title('The Dow Jones Industrial Average (DJIA) from April 20, 2006 to April 20, 2016')
plt.xlabel('Time')
plt.ylabel('DJIA')
plt.show()
```

Requirements for stationarity: (i) First order moment E[x(t)] is independet of time t and therefore constant; (ii) Autocovariance function is independent of time t.
Based on the plotted time series, the expected value of the close index is not constant and therefore the time series cannot be stationary. 

(b)

```python
import numpy as np
daily_returns = np.diff(np.log(df['Close']), n=1)
time = df.index.to_numpy()
time = np.delete(time, 0)
```

```python
plt.plot(time, daily_returns)
plt.title('The daily returns of the Dow Jones Industrial Average (DJIA) from April 20, 2006 to April 20, 2016')
plt.xlabel('Time')
plt.ylabel('DJIA Returns')
plt.show()
```

Based on the same stationarity requirements defined before, one can observe that returns appear to have a constant expected value which does not depend on time. Despite some outlying observations during financial crises/booms (2008-2009, 2010 and 2015), the variance is also quite constant over time. Therefore, in accordance with the empirics, one can conclude that asset returns are much closer to stationary time series than asset prices.

```python
n = 200
arr = []
for t in range(1,n+1):
    xt = t
    arr = np.append(arr,xt)
plot_acf(arr, lags = 10)
```

```python
n = 900
arr = []
for t in range(1,n+1):
    xt = t
    arr = np.append(arr,xt)
plot_acf(arr, lags = 10)
```

The autocorrelation of lag 10 increases for higher n's, which underlines the findings of part 1 that the ACF(h) converges to 1 as n goes to infinity.

(c)

```python
from random import gauss
n = 600
wt = [gauss(0.0, 1.0) for i in range(n)]
wt = np.array(wt)
xt = []
for i in range(n):
    x = wt[i] + 0.5*wt[i-1]
    xt = np.append(xt, x)
time = np.arange(n)
plt.scatter(time, xt)
plt.title('x(t)')
plt.xlabel('Time')
plt.show()
plot_acf(xt)
```

Based on the moving average termination of x(t), one can analogously observe that the ACF shows serial dependence until lag 1 (as can be seen in the moving average formula as well). After lag 1, the correlation is inside the bounds and therefore not significant anymore. The other way around, one can conclude that a ACF of a time series with an unknown stochastic process with a serial dependence up to lag 1 could come from a moving average termination of order 1. Therefore, one could test to approximate the time series with a moving average time series or order 1.
