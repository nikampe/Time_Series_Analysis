### Part I. Augmented Dickey-Fuller (ADF)

#### Causal AR(2)

```python
# generate n observations of AR(1)
phi =   np.array([0.7, 0.2])
theta = np.array([0.])
         
ar = np.r_[1, -phi]
ma = np.r_[1, theta]

arma_process = smt.ArmaProcess(ar, ma)
print("Is the process causal?      Answer:",arma_process.isstationary)
print("Is the process invertible?  Answer:",arma_process.isinvertible)
print("\n")
    
np.random.seed(167867)
sigma = 1.0
n=200
xt = smt.arma_generate_sample(ar=ar, ma=ma, sigma=sigma, nsample=n)

plt.figure(figsize=(16,4))
plt.plot(xt)
plt.xlabel("t")
plt.ylabel("$x_t$")
plt.title("Causal AR(2)")

print('p-value of the Augmented Dickey-Fuller test: %f' % adfuller(xt)[1])
```

#### Non-Causal AR(2)

```python
# generate n observations of AR(1)
phi =   np.array([1.7, -0.7])
theta = np.array([0.])
         
ar = np.r_[1, -phi]
ma = np.r_[1, theta]

arma_process = smt.ArmaProcess(ar, ma)
print("Is the process causal?      Answer:",arma_process.isstationary)
print("Is the process invertible?  Answer:",arma_process.isinvertible)
print("\n")
    
np.random.seed(167867)
sigma = 1.0
n=200
xt = smt.arma_generate_sample(ar=ar, ma=ma, sigma=sigma, nsample=n)

plt.figure(figsize=(16,4))
plt.plot(xt)
plt.xlabel("t")
plt.ylabel("$x_t$")
plt.title("Non-Causal AR(2)")

print('p-value of the Augmented Dickey-Fuller test: %f' % adfuller(xt)[1])
```

### Part II. Feedforward Fully Connected Neural Networks

Let's use Neural Networks (NNs) for modeling $E[x_t|x_{t-1},x_{t-2},\ldots]$ directly, i.e. without a time series model. Let's consider an example of time series with $E[x_t|x_{t-1},x_{t-2},\ldots] = -0.7 x_{t-1}$:

```python
# Generate AR(1) with phi = -0.7
phi = np.array([-0.7])
theta = np.array([0.])

ar = np.r_[1, -phi]
ma = np.r_[1, theta]
sigma = 1

n=20000
np.random.seed(123456)
xt = smt.arma_generate_sample(ar=ar, ma=ma, sigma=sigma, nsample=n)
```

We can see from the ACF / PACF plots that AR(1) is a good candidate for the time series model of this data:

```python
h_max = 50

fig, ax = plt.subplots(nrows=1, ncols=2, sharex=False, figsize=(15,4))
#fig.suptitle('Sample ACF and PACF of ARMA(2) with $\phi_1=$'+str(phi[0])+", $\\theta_1=$"+str(theta[0]),y=1.08,size=16)
plt.subplots_adjust(wspace = 0.15)
plot_acf(xt, ax=ax[0], lags=h_max)
plot_pacf(xt, ax=ax[1], lags=h_max, method='ywm')

for k in range(2):
    ax[k].tick_params(axis='both', labelsize=12)
    [ax[k].spines[line].set_alpha(.3) for line in ["top","bottom","right","left"]]

ax[0].set_title("Sample ACF")
ax[1].set_title("Sample PACF")

plt.show()
```

Let's try fitting AR(1) then:

```python
# estimate parameters of AR(1) based on the sample run xt of n observations
p = 1
model = ARIMA(xt, order=(p,0,0))
model_fit = model.fit(method='css')
#print(model_fit.summary())

# check n fitted values
xt_fitted = model_fit.predict()
plt.figure(figsize=(16,6))
plt.plot(range(1,n+1),xt,linestyle=':', marker='o',color='blue',label = "$x_t$")
plt.plot(range(p+1,n+1),xt_fitted,linestyle=':', marker='o', mfc='none', color='red',label = "1-step ahead predictions: AR")

# forecast m steps ahead starting from n+1
m=10
xt_forecasted = model_fit.predict(start=n+1, end=n+m-1)
xt_forecasted
plt.axvline(x=n,linestyle=':',color='k')
plt.plot(range(n+1,n+m+1), xt_forecasted, linestyle=':', marker='o',color='red',label = "forecasted $x_t$: AR")
plt.title("AR(1) with $\phi = $" + str(phi[0]))
plt.legend()

plt.xlim(n-50,n+m+1)
plt.ylim(-3.5,3.5)
```

We now assume that we do not know whether $E[x_t|x_{t-1},x_{t-2},\ldots]$ is a linear function and want to model it with a NN. We need to scale the data first as the activation functions have specific ranges which may not cover the original data points:

```python
#rescale the time series
xt_min = min(xt)
xt_max = max(xt)
xt = (xt-xt_min)/(xt_max-xt_min)
```

Next, we create a data frame suitable for supervised learning algorithms (in this case NN for regression). In this example, we assume $E[x_t|x_{t-1},x_{t-2},\ldots]$ depends on $x_{t-1}$ only (generally, we could take more lags and/or even skipping some of them like in the case of seasonal processes):

```python
xt = pd.Series(xt)

data = concat([xt.shift(1), xt], axis=1)
data.columns = ['x(t-1)', 'x(t)']
display(data.head())
```

NNs have many parameters and therefore are prone to overfitting data. We need to make sure we properly validate the model and hence we need to split the time series into train and test parts:

```python
# split into train and test sets
n_train = int(n * 0.80)
n_test = n - n_train
#print("n_train = ",str(n_train))
#print("n_test = ",str(n_test))

train, test = data.iloc[0:n_train,:], data.iloc[n_train:n,:]
#print("\nOriginal data:" )
#display(train.head())

#print("\nNAs removed:" )
train = train.dropna(inplace=False)
#display(train.head())


p = data.shape[1]-1
X_train, y_train = train.iloc[:,0:p], train.iloc[:,p]
X_test, y_test = test.iloc[:,0:p], test.iloc[:,p]

#print(str(p) + " features are used:")
#display(X_train.head())
```

We are ready to create the model. Let's use fully connected deep NN with two hidden layers with 69 parameters:

```python
p = X_train.shape[1]

model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(p,)))
model.add(layers.Dense(2, activation='relu'))
model.add(layers.Dense(1, activation='linear')) 

model.summary()
```

Let's use 'adam' optimization algorithm and minimize the MSE. On every iteration, we will use 2,000 observations (couples ($x_{t-1}, x_t$) in this example); the total number of epochs, i.e. the number of times the entire data set will be used, is set to 600. All these parameters, including the NN architecture itself, actually need to be chosen experimentally, by minimizing the test MSE (i.e. the validation error).

```python
nepochs = 600
model.compile(optimizer='adam',
              loss='mse')
history = model.fit(X_train, y_train,
          epochs=nepochs,
          batch_size=2000,
          shuffle=False,
          validation_data=(X_test, y_test))

history_dict = history.history
```

Below we plot the train and test errors versus the number of epochs. We can see that we could actually stop around 140. In some cases, and this is very typical for NNs, test MSE may go up and therefore it is important to stop early then, but not in this case: 

```python
plt.clf()   # clear figure
train_mse = history_dict['loss']
test_mse = history_dict['val_loss']

plt.plot(range(1,nepochs+1), train_mse, 'b', label='train MSE',color="blue")
plt.plot(range(1,nepochs+1), test_mse, 'b', label='test MSE',color="red")

#plt.ylim((0, 0.0001))
plt.title('min(test MSE) = ' + str(round(min(test_mse),6)))
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.legend()

plt.show()
```

Finally, we are ready to use the trained NN: 

```python
xt_fitted_nn = model.predict(data.iloc[:,0:1])

m = 10
xt_forecasted_nn = np.empty(m+p)
xt_forecasted_nn[0:p] = xt[(n-p):n] 
for t in range(p,m+p):
    xt_forecasted_nn[t] = model.predict([xt_forecasted_nn[t-1]])
xt_forecasted_nn = xt_forecasted_nn[p:(n+p)]

xt = xt_min + xt*(xt_max-xt_min)
xt_fitted_nn = xt_min + xt_fitted_nn*(xt_max-xt_min)
xt_forecasted_nn = xt_min + xt_forecasted_nn*(xt_max-xt_min)

# check n fitted values
#xt_fitted_nn = model.predict(data.iloc[:,0:1])
#xt_fitted_nn = xt_min + xt_fitted_nn*(xt_max-xt_min)
plt.figure(figsize=(16,6))
plt.plot(range(1,n+1),xt,linestyle=':', marker='o',color='blue',label = "$x_t$")
plt.plot(range(1,n+1),xt_fitted_nn,linestyle=':', marker='o', mfc='none', color='green',label = "1-step ahead predictions: NN")

# forecast m steps ahead starting from n+1
#m = 10
#xt_forecasted_nn = np.empty(m+p)
#xt_forecasted_nn[0:p] = xt[(n-p):n] 
#for t in range(p,m+p):
#    xt_forecasted_nn[t] = model.predict([xt_forecasted_nn[t-1]])
#xt_forecasted_nn = xt_forecasted_nn[p:(n+p)]

plt.axvline(x=n,linestyle=':',color='k')
plt.plot(range(n+1,n+m+1), xt_forecasted_nn, linestyle=':', marker='o',color='green',label = "forecasted $x_t$: NN")
plt.title("AR(1) with $\phi = $" + str(phi[0]))
plt.legend()

plt.xlim(n-50,n+m+1)
plt.ylim(-3.5,3.5)
```

We see that the result of fitting AR(1) and the NN are nearly identical in this example. We notice that the 1-step ahead predictions obtained by the NN are actually obtained for new points because the NN was trained on the first 80% of observations and did not use the last 20% of time series data:

```python
# check n fitted values
plt.figure(figsize=(16,6))
plt.plot(range(1,n+1),xt,linestyle=':', marker='o',color='blue',label = "$x_t$")
plt.plot(range(p+1,n+1),xt_fitted,linestyle=':', marker='o', mfc='none', color='red',label = "1-steap ahead predictions: AR")
plt.plot(range(1,n+1),xt_fitted_nn,linestyle=':', marker='o', mfc='none', color='green',label = "1-step ahead predictions: NN")

# forecast m steps ahead starting from n+1
plt.axvline(x=n,linestyle=':',color='k')
plt.plot(range(n+1,n+m+1), xt_forecasted, linestyle=':', marker='o',color='red',label = "forecasted $x_t$: AR")
plt.plot(range(n+1,n+m+1), xt_forecasted_nn, linestyle=':', marker='o',color='green',label = "forecasted $x_t$: NN")
plt.title("AR(1) with $\phi = $" + str(phi[0]))
plt.legend(loc='upper left')

plt.xlim(n-50,n+m+1)
plt.ylim(-3.5,3.5)
```

### 2. Time Series Forecasting via RNN

#### (a) Generating Time Series:
Let's generate a time series with non-linear $E[x_t|x_{t-1},x_{t-2},\ldots]$:

```python
def ts_simulate(phi,phi_interaction,omega,alpha,beta,n=100):
    
    n0 = 200
    n = n0 + n
    
    m = len(phi)
    p = len(alpha)
    q = len(beta)
        
    sigma = np.zeros(n)
    rt = np.zeros(n)
    xt = np.zeros(n)
    epsilon = np.random.normal(0, 1, n)
    
    for t in range(np.max([m, p, q])):
        sigma[t] = np.random.normal(0, 1)
        rt[t] = sigma[t] * epsilon[t]
        xt[t] = rt[t]
        
    for t in range(np.max([m, p, q]), n):
        term1 = omega + sum([alpha[i] * rt[t-1-i]**2 for i in range(p)])
        term2 = sum([beta[i] * sigma[t-1-i]**2 for i in range(q)])
        sigma[t] = np.sqrt(term1 + term2)
        rt[t] = sigma[t] * epsilon[t]
        xt[t] = phi[0] * xt[t-1] + phi[1] * xt[t-2] + phi_interaction * xt[t-1] * xt[t-2]  + rt[t]
        
    return xt[n0:]
```

```python
np.random.seed(54323)
phi = np.array([0.9,-0.2])
phi_interaction = 0.02
omega = 0.01
alpha = np.array([0.3, 0.6])
beta = np.array([0])

p_ar = 0 if ((len(phi)==1) & (phi[0]==0)) else len(phi)
p_garch = 0 if ((len(alpha)==1) & (alpha[0]==0)) else len(alpha)
q_garch = 0 if ((len(beta)==1) & (beta[0]==0)) else len(beta)

n = 20000
xt = ts_simulate(phi,phi_interaction,omega,alpha,beta,n)
xt = (xt-xt.min())/(xt.max()-xt.min())

plt.figure(figsize=(16,6))
plt.plot(xt, linestyle='--', marker='o', markersize=4)
plt.xlabel("Year")
plt.ylabel("$x_t$")
plt.title("Time series with non-linear $E[x_t|x_{t-1},x_{t-2},\ldots]$")
plt.grid(True)  

plt.figure(figsize=(16,6))
plt.plot(range(1,n+1),xt, linestyle='--', marker='o', markersize=4)
plt.xlabel("Year")
plt.ylabel("$x_t$")
plt.title("Time series with non-linear $E[x_t|x_{t-1},x_{t-2},\ldots]$")
plt.axvline(x=n,linestyle=':',color='k')
plt.grid(True)
plt.xlim(n-140,n+10)
plt.ylim(0.2,0.86)
```

#### (b) Preparing data for RNNs:

```python
n_timesteps = 14

features_set = []
labels = []
for i in range(n_timesteps, xt.shape[0]):
    features_set.append(xt[i-n_timesteps:i])
    labels.append(xt[i])
    
features_set, labels = np.array(features_set), np.array(labels)
features_set = np.reshape(features_set,(features_set.shape[0], features_set.shape[1], 1))

#inputs: A 3D tensor with shape [batch, timesteps, feature]
```

```python
print("features_set.shape is ", features_set.shape)
print("labels.shape is ", labels.shape)
```

```python
# split into train and test sets
n = features_set.shape[0]
n_train = int(n * 0.80)
n_test = n - n_train
#print("n_train = ",str(n_train))
#print("n_test = ",str(n_test))

features_set_train, features_set_test =  features_set[0:n_train,:,:], features_set[n_train:n,:,:]
labels_train, labels_test =  labels[0:n_train], labels[n_train:n]

print("features_set_train.shape is ", features_set_train.shape)
print("labels_train.shape is ", labels_train.shape)
print("\nfeatures_set_test.shape is ", features_set_test.shape)
print("labels_test.shape is ", labels_test.shape)
```

#### (c) Time Series Forecasting via Simple RNN with 6 Recurrent Neurons

```python
n_features = 1

model = models.Sequential()
model.add(layers.SimpleRNN(6, activation='relu', input_shape=(n_timesteps,n_features)))
model.add(layers.Dense(1, activation='linear')) 

model.summary()
```

```python
nepochs = 40
model.compile(optimizer='adam',
              loss='mse')
history = model.fit(features_set_train, labels_train,
          epochs=nepochs,
          batch_size=32,
          validation_data=(features_set_test, labels_test))
history = history.history
```

```python
plt.clf()   # clear figure
train_mse = history['loss']
test_mse = history['val_loss']

plt.figure(figsize=(16,6))
plt.plot(range(1,nepochs+1), train_mse, 'b', label='train MSE',color="blue")
plt.plot(range(1,nepochs+1), test_mse, 'b', label='test MSE',color="red")

plt.title('Simple RNN: min(test MSE) = ' + str(round(min(test_mse),6)),fontsize=16)
plt.xlabel('Epochs',fontsize=16)
plt.ylabel('MSE',fontsize=16)
plt.legend(loc = 'upper right',fontsize=16)
#plt.ylim(0,0.001)

plt.show()
```
