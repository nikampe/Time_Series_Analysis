## Example 3.2 The Sample Path of an AR(1) Process

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.tsa.api as smt

from statsmodels.graphics.tsaplots import plot_acf 
from statsmodels.tsa.stattools import acf
```

#### AR(1) with $\phi = 0.9$

```python
# AR(1) with phi = 0.9
np.random.seed(123456789)
n = int(100)

phi = np.array([0.9])
theta = np.array([0.])

ar = np.r_[1, -phi]
ma = np.r_[1, theta]
sigma = 1
print("coefficients of the autoregressive operator:", ar)
print("coefficients of the moving average operator:", ma)

xt = smt.arma_generate_sample(ar=ar, ma=ma, sigma=sigma, nsample=n) 

plt.figure(figsize=(16,4))
plt.plot(xt)
plt.ylabel("$x_t$",fontsize=14)
plt.title("AR(1) with $\phi=+0.9$")
plt.grid(True)
```

```python
fig = plot_acf(xt,lags = 20)
plt.xlabel("lag")
plt.ylabel("ACF")
plt.title("ACF of AR(1) with $\phi=+0.9$")
```

```python
fig, (h1, h2, h3, h4) = plt.subplots(nrows=1, ncols=4, sharex=False, figsize=(15,3))
fig.suptitle('AR(1) with $\phi=+0.9$',y=1.08,size=16)
plt.subplots_adjust(wspace = 0.35)

h1.plot(xt[range(0,n-1)],xt[range(1,n)],linestyle="None", marker='o', color='k')
h1.set_title("lag = 1")
h1.set_xlabel("$x_{t-1}$",fontsize=14)
h1.set_ylabel("$x_t$",fontsize=14)

h2.plot(xt[range(0,n-2)],xt[range(2,n)],linestyle="None", marker='o', color='k')
h2.set_title("lag = 2")
h2.set_xlabel("$x_{t-2}$",fontsize=14)
h2.set_ylabel("$x_t$",fontsize=14)

h3.plot(xt[range(0,n-3)],xt[range(3,n)],linestyle="None", marker='o', color='k')
h3.set_title("lag = 3")
h3.set_xlabel("$x_{t-3}$",fontsize=14)
h3.set_ylabel("$x_t$",fontsize=14)

h4.plot(xt[range(0,n-4)],xt[range(4,n)],linestyle="None", marker='o', color='k')
h4.set_title("lag = 4")
h4.set_xlabel("$x_{t-4}$",fontsize=14)
h4.set_ylabel("$x_t$",fontsize=14)

plt.show()
```
