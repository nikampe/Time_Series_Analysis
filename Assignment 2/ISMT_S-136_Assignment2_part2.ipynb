## Problem 1 (10 points)
In this problem, we numerically investigate stationarity of $x_t=\sin(2\pi Ut)$, $t=1,2,3,\ldots$, where $U\sim \text{Unif}(0,1)$.  

(a) Generate $U$ from a uniform distribution on the interval $(0,1)$ 4 times. For each $U$, plot the corresponding $x_t$, $t=1,2,\ldots,20$ in one figure.    

(b) Generate $U$ from a uniform distribution on the interval $(0,1)$ 1000 times. For each $U$, generate $x_1, x_2, x_3, x_4$. Plot the 2D histograms of $(x_1,x_3)$ and $(x_2,x_4)$. You can use plt.hist2d. Discuss whether $x_1$ appears to be strictly stationary.

SOLUTION:

(a)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

np.random.seed(453)

t = np.array(range(1,21))
for i in range(4):
    x = []
    Ut = np.random.uniform(low = 0.0, high = 1.0, size = 1)
    for j in range(20):
        xt = np.sin(2 * math.pi * Ut * j)
        x = np.append(x, xt)
    plt.scatter(t, x)
    plt.title('x(t)')
    plt.xlabel('Time')
    plt.show()
```

(b)

```python
np.random.seed(453)

columns = np.array(range(1,1001))
index = np.array(range(1,5))
df = pd.DataFrame(columns = columns, index = index)

for i in range(1,1001):
    Ut = np.random.uniform(low = 0.0, high = 1.0, size = 1)
    x = []
    for j in range(1,5):
        xt = np.sin(2 * math.pi * Ut * j)
        x = np.append(x, xt)
    df[i] = x
plt.hist2d(df.loc[1,:], df.loc[3,:])
plt.show()
plt.hist2d(df.loc[2,:], df.loc[4,:])
plt.show()
```

x(1) appears to be strictly stationary because it always has the same joint distribution at any lag/for any U(t) as U(t) is created from a uniformed distribution where the 1000 variables are uncorrelated and identically distributed. Therefore, every value in x(1) has the same underlying stochastic properties and hence x(1) is strictly stationary.

SOLUTION:

(a)

```python
df = pd.read_csv('chicken.csv', index_col = 'date')

plt.plot(df.index, df['chicken'])
plt.title('The price of chicken: monthly whole bird spot price, Georgia docks, US cents per pound, August 2001 to July 2016, with fitted linear trend line')
plt.xlabel('Time')
plt.ylabel('Cents per pound')
plt.show()
```

The time series of prices of chicken does not seem to be stationary as one can clearly observe that the mean is not constant over time, but either increasing in t. Therefore, the requirement of weak stationarity that the mean is independent of t is not given and we must reject stationarity.

(b)

```python
from scipy import signal
from sklearn.linear_model import LinearRegression

y = df['chicken'].to_numpy().reshape(1, -1)
x = df.index.to_numpy().reshape(1, -1)

regr = LinearRegression()
regr.fit(x, y)
y_pred = regr.predict(x)

plt.scatter(x, y, color = 'black')
plt.plot(x, y_pred, color = 'blue', linewidth = 3)
plt.show()
```

Somehow the graph does not show the fitted linear trend.

(c)

```python
chicken_diff = []
for i in range(len(df['chicken'].to_numpy())):
    chicken_diff_t = np.nan
    chicken_diff = np.append(chicken_diff, chicken_diff_t)
df['chicken_diff'] = chicken_diff
df.iloc[0, 1] = 0
for i in range(len(df['chicken'].to_numpy()) - 1):
    df.iloc[i+1, 1] = df.iloc[i+1, 0] - df.iloc[i, 0]
plt.plot(df.index, df['chicken_diff'])
plt.show()
```

The first difference of the chicken time series must fulfill a time-independent mean and a time-independent variance. From the plot, one can observe that the mean might be constant at around zero and thus independent of time. The variance looks almost independent of t but seems to decrease over time. Therefore the first difference is almost stationary, but has a variance which is not perfectly independent of t.
